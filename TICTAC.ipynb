```python
# creatign the display
def print_board(board):
    print('\n'*6)
    print(board[1] + '   |   '+ board[2] +'   |   ' + board[3])
    print("--------------------")
    print(board[4] + '   |   '+ board[5] +'   |   ' + board[6])
    print("--------------------")
    print(board[7] + '   |   '+ board[8] +'   |   ' + board[9])
    
#assigning X and O
def player_assign():
    assignment=''
    while assignment!='X' and assignment!='O':
        assignment=input("player 1 please choose either X or O: ")
        if assignment!='X' and assignment!='O':
            print("invalid response: please enter X/O")
        if assignment=='X':
            player1="X"
            player2="O"
        else:
            player1="O"
            player2="X"
    return (player1, player2)
    
    
 # game continutaion
def gameon_choice():
    ans=''
    while ans!='N' and ans!='Y':
        ans=input("DO you want to continue playing enter Y/N")
        if ans!='N' and ans!='Y':
            print("invalid respsonse please enter Y/N")
        elif ans=="Y":
            return True
        else:
            return False
           
# getting user input          

def get_user_input(game_board):
    position=''
    pos_range=['1','2','3','4','5','6','7','8','9']
    print_board(game_board)
    while position not in pos_range:
        position=input("please enter your position:1-9 ")
        if position not in pos_range:
            print("invalid position: please enter correct value") 
        elif game_board[int(position)]!='':
            print("already a value present please select another position")
            position=''
    return int(position)

#check in game over

def game_over(game_board):
    if   game_board[1]==game_board[2]==game_board[3]=="O" or game_board[1]==game_board[2]==game_board[3]=="X":
        return True
    elif game_board[4]==game_board[5]==game_board[6]=="O" or game_board[4]==game_board[5]==game_board[6]=="X":
        return True
    elif game_board[7]==game_board[8]==game_board[9]=="O" or game_board[7]==game_board[8]==game_board[9]=="X":
        return True
    elif game_board[1]==game_board[4]==game_board[7]=="O" or game_board[1]==game_board[4]==game_board[7]=="X":
        return True
    elif game_board[2]==game_board[5]==game_board[8]=="O" or game_board[2]==game_board[5]==game_board[8]=="X":
        return True
    elif game_board[3]==game_board[6]==game_board[9]=="O" or game_board[3]==game_board[6]==game_board[9]=="X":
        return True
    elif game_board[1]==game_board[5]==game_board[9]=="O" or game_board[1]==game_board[5]==game_board[9]=="X":
        return True
    elif game_board[3]==game_board[5]==game_board[7]=="O" or game_board[3]==game_board[5]==game_board[7]=="X":
        return True
    else:
        return False

#MAIN PROGRAM
game_on=True
temp_board=['','1','2','3','4','5','6','7','8','9']
game_board=['']*10
game_board[0]='dummy'
probe=True
choice=''
counter=1
empty=True
winner=False
print(" lets start the game")
print( " below are the postion in the board")
print_board(temp_board)
while game_on:
    player1,player2=player_assign()
    print_board(game_board)
    
    while empty and  winner==False:
        if counter==1:
            print (" player1 turn")
            choice=get_user_input(game_board)
            game_board[choice]=player1
            counter=2
        else:
            print("player 2 turn")
            choice=get_user_input(game_board)
            game_board[choice]=player2
            counter=1
        winner=game_over(game_board)
        if '' not in game_board:
            empty=False
    print_board(game_board)        
    if winner and counter==1:
        print("player 2 wins")
    elif winner and counter==2:
        print("player 1 wins") 
    else:
        print("game is a draw")
    game_on=gameon_choice()
```
